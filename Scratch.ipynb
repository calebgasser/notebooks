```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
fig, axs = plt.subplots(3)

x1 = np.linspace(-100, 100, 100)
y1 = 1 / (1 + np.exp(-x1))

w = np.random.normal(50, 1, (100,100))
b = 0.0
x2 = np.linspace(-100, 100, 100)
y2 = np.dot(x2, w.T) + b

x3 = x1
y3 = 1 / (1 + np.exp(-x3))

axs[0].plot(x2, y2)
axs[0].set_title("Weights")
axs[0].set_xlabel("x values")
axs[0].set_ylabel("Weights of (x)")

axs[1].plot(x1, y1)
axs[1].set_title("Sigmoid Function")
axs[1].set_xlabel("x values")
axs[1].set_ylabel("Sigmoid of (x)")

axs[2].plot(x3, y3)
axs[2].set_title("Sigmoid Function to weights")
axs[2].set_xlabel("x values")
axs[2].set_ylabel("Sigmoid of (x)")
```
